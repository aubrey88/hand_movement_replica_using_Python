```python
resses = cv2.imread('ComputerVisionCourse/DATA/reeses_puffs.png',0)

```

```python
plt.imshow(resses)
```

```python
cereals = cv2.imread('ComputerVisionCourse/DATA/many_cereals.jpg',1)

```

```python
plt.imshow(cereals)

```

```python
#create a detectorr obj for feature mapping

orb = cv2.ORB_create()
```

```python
kb1,des1= orb.detectAndCompute(resses, None)
kb2,des2= orb.detectAndCompute(cereals, None)

#
```

```python
#brute force matching to match the features of 2 images

bf= cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True )

matches= bf.match(des1,des2)
```

```python
matches[:30]
```

```python
#finding the matching percentage of the first element in marches

single_match= matches[0]

```

```python
matches= sorted(matches,key=lambda x:x.distance)

```

```python
#none here means no masking
reeses_matches=cv2.drawMatches(resses,kb1,cereals,kb2,matches[:10],None, flags=2)
```

```python
plt.imshow(reeses_matches)
```

```python
sift= cv2.xfeatures2d.SIFT_create()

kp1,des1= sift.detectAndCompute(resses, None)

kp2,des2= sift.detectAndCompute(cereals, None)

```

```python
bf= cv2.BFMatcher()

matches= bf.knnMatch(des1,des2,k=2)
```

```python
matches[:10]
```

```python
good= []

for match1, match2 in matches:

    #if match 1 distance is less than 75% of match 2 distance

#then  des

    if match1.distance < 0.75*match2.distance:

                      good.append([match1])

```

```python
#check the new length of good/ successful matches

len(good)
```

```python
display(good)
```

```python
#none here means no masking
swift_matches=cv2.drawMatchesKnn(resses,kp1,cereals,kp2,good,None, flags=2)
```

```python

def display(param):
    plt.imshow(param)
```

```python
display(swift_matches)
```

```python
#FLANN is not as accurate as brute force but it is faster 

FLANN_INDEX_KDTREE=0

index_params = dict (algorithm = FLANN_INDEX_KDTREE, trees = 5)

search_params = dict(checks=50)

```

```python
flann= cv2.FlannBasedMatcher(index_params, search_params)

```

```python
matches = flann.knnMatch(des1,des2,k=2)
```

```python
good_matches_using_Flann= []

for match1, match2 in matches:

    #if match 1 distance is less than 75% of match 2 distance

#then  des

    if match1.distance < 0.75*match2.distance:

                      good_matches_using_Flann.append([match1])
```

```python
flann_matches=cv2.drawMatchesKnn(resses,kp1,cereals,kp2,good_matches_using_Flann,None, flags=0) 
```

```python
display(flann_matches)
```

```python
#face detection lecture 54

nadia = cv2.imread('ComputerVisionCourse/DATA/Nadia_Murad.jpg', 0)

denis = cv2.imread('ComputerVisionCourse/DATA/Denis_Mukwege.jpg', 0)

solway = cv2.imread('ComputerVisionCourse/DATA/solvay_conference.jpg',2)

plt.imshow(solway, cmap='gray')

face_cascade = cv2.CascadeClassifier('ComputerVisionCourse/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
# creating a detect face function 

def detect_face(img):

         face_img = img.copy()

         face_rects = face_cascade.detectMultiScale(face_img)

         for (x,y, w, h) in face_rects:

             cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)

         return face_img
```

```python
result = detect_face(solway)
```

```python
# creating a detect face function

def adj_detect_face(img):

    face_img = img.copy()

    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)

    for (x,y, w, h) in face_rects:

     cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)

    return face_img
```

```python
adjacent_result= adj_detect_face(solway)
```

```python
plt.imshow(adjacent_result, cmap='gray')
```

```python
eye_cascade= cv2.CascadeClassifier('ComputerVisionCourse/DATA/haarcascades/haarcascade_eye.xml')

#function to detect eyes in faces

def detect_eye(img):

    face_img = img.copy()

    eye_rects = eye_cascade.detectMultiScale(face_img)

    for (x,y, w, h) in eye_rects:

     cv2.rectangle(face_img, (x,y), (x+w, y+h), (255,255,255), 10)

    return face_img
```

```python
eyedetect = detect_eye(solway)
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

lower_skin = np.array([0, 20, 70], dtype=np.uint8)
upper_skin = np.array([20, 255, 255], dtype=np.uint8)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    mirrored_frame = cv2.flip(frame, 1)

    hsv = cv2.cvtColor(mirrored_frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, lower_skin, upper_skin)

    contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    gray_mirrored = cv2.cvtColor(mirrored_frame, cv2.COLOR_BGR2GRAY)
    bw_mirrored = np.zeros_like(mirrored_frame)

    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(mirrored_frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        bw_mirrored[y:y+h, x:x+w] = cv2.cvtColor(mirrored_frame[y:y+h, x:x+w], cv2.COLOR_BGR2GRAY)[:, :, None]

    cv2.imshow('Original Frame', mirrored_frame)
    cv2.imshow('Black and White Mirror Frame', bw_mirrored)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
